```python
# Drop duplicates
df.drop_duplicates(inplace=True)
```

```python
df["Job Title"].unique()
```

```python
# Clean job titles
df["Job Title"] = df["Job Title"].str.strip().str.lower()
```

4. Outliers check

```python
df["Salary Estimate"].describe()
```

```python
df["Salary Estimate"].unique()
```

```python
# clean sallary range
df["Salary Estimate"] = df["Salary Estimate"].apply(lambda x: x.split('(')[0])
```

```python
df["Salary Estimate"] = df["Salary Estimate"].apply(lambda x: x.replace('K', '').replace('$', ''))
```

```python
df["min_salary"] = df["Salary Estimate"].apply(lambda x: int(x.split('-')[0]))
df["max_salary"] = df["Salary Estimate"].apply(lambda x: int(x.split('-')[1]))
df.head()
```

```python
# Remove extreme salary outliers
q1 = df['min_salary'].quantile(0.25)
q3 = df['min_salary'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr
df = df[(df['min_salary'] >= lower_bound) & (df['min_salary'] <= upper_bound)]
```

```python
# Remove extreme salary outliers
q1 = df['max_salary'].quantile(0.25)
q3 = df['max_salary'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr
df = df[(df['max_salary'] >= lower_bound) & (df['max_salary'] <= upper_bound)]
```

```python
df.info()
```

## Data Preprocessing

```python
df.dtypes
```

```python
df["avg_salary"] = (df.min_salary + df.max_salary)//2
df.head()
```

```python
df["Company Name"] = df["Company Name"].apply(lambda x: x.split('\n')[0])
df.head()
```

```python
df.sort_values(by="Rating", ascending=True)
```

```python
df.Rating = np.where(df.Rating == -1.0, 0, df.Rating)
```

```python
df.sort_values(by="Rating", ascending=True)
```

```python
df["job_state"] = df["Location"].apply(lambda x: x.split(",")[-1])
df.head()
```

```python
df["same_state"] = df.apply(lambda x: 1 if x.Location == x.Headquarters else 0, axis=1)
df.head()
```

```python
df.Founded
```

```python
df["company_age"] = df.Founded.apply(lambda x: x if x < 1 else 2025-x)
df.head()
```

```python
df.company_age = np.where(df.company_age == -1.0, 0, df.company_age)
df[df["company_age"] == -1].shape
```

```python
df["Job Description"][2].split('\n')
```

```python
df['python'] = df['Job Description'].apply(lambda x: 1 if 'python' in x.lower() else 0)
df['excel'] = df['Job Description'].apply(lambda x: 1 if 'excel' in x.lower() else 0)
df['R'] = df['Job Description'].apply(lambda x: 1 if ' r ' in x.lower() else 0)
df['NLP'] = df['Job Description'].apply(lambda x: 1 if 'nlp' in x.lower() else 0)
df['hadoop'] = df['Job Description'].apply(lambda x: 1 if 'hadoop' in x.lower() else 0)
df['spark'] = df['Job Description'].apply(lambda x: 1 if 'spark' in x.lower() else 0)
df['aws'] = df['Job Description'].apply(lambda x: 1 if 'aws' in x.lower() else 0)
df['tableau'] = df['Job Description'].apply(lambda x: 1 if 'tableau' in x.lower() else 0)
df['big_data'] = df['Job Description'].apply(lambda x: 1 if 'big data' in x.lower() else 0)
df.head()
```

```python
def title_simplifier(title):
    if 'data scientist' in title.lower():
        return 'data scientist'
    elif 'data engineer' in title.lower():
        return 'data engineer'
    elif 'analyst' in title.lower():
        return 'analyst'
    elif 'machine learning' in title.lower():
        return 'mle'
    elif 'manager' in title.lower():
        return 'manager'
    elif 'director' in title.lower():
        return 'director'
    else:
        return 'na'
    
def seniority(title):
    if 'sr' in title.lower() or 'senior' in title.lower() or 'sr' in title.lower() or 'lead' in title.lower() or 'principal' in title.lower():
            return 'senior'
    elif 'jr' in title.lower() or 'jr.' in title.lower():
        return 'jr'
    else:
        return 'entry level'

df['job_simp'] = df['Job Title'].apply(title_simplifier)

df.job_simp.value_counts()
```

```python
df["seniority"] = df["Job Title"].apply(seniority)
df.seniority.value_counts()
```

```python
df.job_state.value_counts()
```

```python
df = df[~((df['job_state']=='Remote') | (df['job_state']=='Utah') | (df['job_state']=='New Jersey') | (df['job_state']=='Texas') | (df['job_state']=='California'))]
df.job_state = df.job_state.apply(lambda x: x.replace('United States','US'))
df.job_state.value_counts()
```

```python
# df.drop(["Founded", "Competitors"], axis=1, inplace=True)
df = df.apply(lambda x: x.strip() if isinstance(x, str) else x)
```

## Data Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize = (8,5))
sns.countplot(x = "seniority", data=df, order=df["seniority"].value_counts().index)
plt.title('Job Posting by Experience Level')
plt.xlabel('Experience Level')
plt.ylabel('Count')
plt.show()
```

```python
sns.kdeplot(df["min_salary"], label="Min Salary", shade="False", color="blue")
sns.kdeplot(df["max_salary"], label="Max Salary", shade="False", color="green")
sns.kdeplot(df["avg_salary"], label="Avg Salary", shade="False", color="cyan")

plt.title("Min vs Max vs Avg Salary KDE Distribution")
plt.xlabel("Salary Amount")
plt.ylabel("Density")
plt.legend()
plt.show()
```

```python
numeric_cols = df[["min_salary", "max_salary", "avg_salary", "company_age", "Rating"]]
correlation_matrix = numeric_cols.corr()

plt.figure(figsize=(6,4))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Heatmap of Salary & Company Age Features")
plt.show()
```
