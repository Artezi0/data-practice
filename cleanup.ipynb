```python
# Objective: clean data
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Get data
df = pd.read_csv("./warmup-data/archive/Uncleaned_DS_jobs.csv")
df = df.drop_duplicates()
df
```

```python
df['Job Title'] = df['Job Title'].str.title()
df
```

```python
# Format the salary estimate
df["Salary Estimate"] = df["Salary Estimate"].str.replace('(Glassdoor est.)', '')
df["Salary Estimate"] = df["Salary Estimate"].str.replace('$', '')
df["Salary Estimate"] = df["Salary Estimate"].str.replace('K', '')
df
```

```python
# Format size collumn
df["Size"] = df["Size"].str.replace("to", "-")
df["Size"] = df["Size"].str.replace("employees", "-")
df
```

```python
# Format the company name
df["Company Name"] = df["Company Name"].str.replace('\n', '')
df["Company Name"] = df["Company Name"].str.replace('[^a-zA-Z]', '', regex=True)
df
```

```python
df.drop(df[df['Rating'] == -1.0].index, inplace=True)
df.drop(df[df['Founded'] == -1].index, inplace=True)
df.drop(df[df['Competitors'] == "-1"].index, inplace=True)
```

```python
# df = df.drop(columns="index")
df = df.reset_index(drop=True)
```
